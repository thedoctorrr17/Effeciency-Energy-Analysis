# Import Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Data & Check Data

```python
df=pd.read_csv("/content/energy_efficiency_data.csv")
```

```python
df.head()
```

```python
print(df.isnull().sum())
print(df.notnull().sum())
```

# ***Univariate Analysis***


```python
column_descriptions = []

# Iterate through each column in the DataFrame
for col in df.columns:
        # Calculate descriptive statistics using the describe() method
        description = df[col].describe()
        column_descriptions.append((col, description))

# Print the descriptive statistics in a clear format
for col_name, description in column_descriptions:
    print(f"\nColumn: {col_name}")
    print(description)
```
